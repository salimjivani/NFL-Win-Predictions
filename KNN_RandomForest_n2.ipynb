```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
def knneighbor (X,y,n):
  X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42, stratify=y)
  knn = KNeighborsClassifier(n_neighbors=n)
  knn.fit(X_train,y_train)
  # Predicting results using Test data set

  error_rate = []
  # Will take some time
  for i in range(1,40):
 
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
  # plt.figure(figsize=(10,6))
  plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
  markerfacecolor='red', markersize=10)
  plt.title('Error Rate vs. K Value')
  plt.xlabel('K')
  plt.ylabel('Error Rate')

  pred = knn.predict(X_test)
  accuracy_score(pred,y_test)
  print('Accuracy\n',accuracy_score(y_test, pred))
  print('\nKNN confusion matrix\n', confusion_matrix(y_test, pred))
  print('\nKNN classification_report\n', classification_report(y_test, pred))
  return pred

```

```python
# Assign colum names to the dataset
# colnames = ['PassingYards','RunningYards','PenaltyYards','Penalties','Turnovers']

# Read dataset to pandas dataframe
dataset = pd.read_csv("/content/NFL_Grouped_Stats_Dataset_for_Model.csv")
X = dataset[['PassingYards','RunningYards','PenaltyYards','Penalties','Turnovers']]
y = dataset['Win_Lose']
```
